### 1. Graph Construction

```python
import pandas as pd
import networkx as nx
import holoviews as hv
import seaborn as sns
from holoviews import opts
hv.extension('bokeh')
from bokeh.plotting import show
from holoviews.operation.datashader import datashade, bundle_graph
import community.community_louvain as community_louvain


# Setting the default figure size a bit larger
defaults = dict(width=750, height=750, padding=0.1,
                xaxis=None, yaxis=None)
hv.opts.defaults(
    opts.EdgePaths(**defaults), opts.Graph(**defaults), opts.Nodes(**defaults))
```

```python
edges = pd.read_csv('edges.csv')
nodes = pd.read_csv('nodes.csv')
```

```python
edges.head()
```

```python
#I make column names lowercase because that is cool
edges.columns = [w.lower() for w in edges.columns]
nodes.columns = [w.lower() for w in nodes.columns]
```

```python
#I make sure the visualization is readable by making the ids provided in the edges.csv into actual labels instead as we know from nodes.csv 
id_to_label = nodes.set_index('id')['label'].to_dict()

edges['source'] = edges['source'].map(id_to_label)
edges['target'] = edges['target'].map(id_to_label)
```

```python
g = nx.from_pandas_edgelist(edges, source='source', target='target', edge_attr=['weight', 'label', 'movie_id'])
```

```python
centrality_dgr = nx.degree_centrality(g)
centrality_eigen = nx.eigenvector_centrality_numpy(g, weight='weight')
centrality_between = nx.betweenness_centrality(g, weight='weight')
```

```python
nx.set_node_attributes(g, centrality_dgr, 'centrality_dgr')
nx.set_node_attributes(g, centrality_eigen, 'centrality_eigen')
nx.set_node_attributes(g, centrality_between, 'centrality_between')
```

```python
partition = community_louvain.best_partition(g)
nx.set_node_attributes(g, partition, 'community')
nodes = pd.DataFrame.from_dict(dict(g.nodes(data=True)),orient='index')
nodes.groupby('community')['centrality_eigen'].nlargest(5)
```

```python
g_layout = nx.layout.spring_layout(g)
g_plot = hv.Graph.from_networkx(g, g_layout).opts(tools=['hover'], node_color='community')
labels = hv.Labels(g_plot.nodes, ['x', 'y'], 'index')
```

```python
bundled = bundle_graph(g_plot)
```

```python
show(hv.render(bundled * labels.opts(text_font_size='6pt', text_color='white', bgcolor='gray')))
```

### 2. Graph Analysis
Determinate the basic graph indicators:
**Total number of nodes and edges**: to understand the *scale and complexity* of the graph
**Average degree**: giving an average of *connections per node*
**Graph density**:  providing an insight into the *overall connectivity*

```python
# a basic information about the graph
print(f"Graph info:")
print(f"Number of nodes: {g.number_of_nodes()}")
print(f"Number of edges: {g.number_of_edges()}")
print(f"Graph density: {nx.density(g)}") # how close a graph is to being a complete graph
print(f"Graph is directed: {g.is_directed()}")
print(f"Average degree: {sum(dict(g.degree()).values()) / g.number_of_nodes():.2f}")
```
