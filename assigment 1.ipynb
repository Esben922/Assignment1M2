### 1. Graph Construction

```python
import pandas as pd
import networkx as nx
import holoviews as hv
import seaborn as sns
from holoviews import opts
hv.extension('bokeh')
from bokeh.plotting import show
from holoviews.operation.datashader import datashade, bundle_graph
import community.community_louvain as community_louvain

defaults = dict(width=750, height=750, padding=0.1,
                xaxis=None, yaxis=None)
hv.opts.defaults(
    opts.EdgePaths(**defaults), opts.Graph(**defaults), opts.Nodes(**defaults))
```

```python
#defining edges.csv and nodes.csv to be used. I define nodes as nodes_df as i expect to populate "nodes" as node attributes later on
edges = pd.read_csv('edges.csv')
nodes_df = pd.read_csv('nodes.csv')
```

```python
#Can see that we have the columns Source and Target that will make up our edgelist. Furthermore we have "type" indicating that the distinction between the nodes is not meaningful, so we cant infer any information
#about the relationship between the nodes from this alone other than the fact that there exists a connection. We can instead look at "Weight" to find out more of the relation. In this case Weight refers to
#how many cooccurences each characters has, so from this we know a bit more about the relation between the nodes.
edges.head()
```

```python
#just quickly making sure all rows are undirected - they are
edges['Type'].unique()
```

```python
#I make column names lowercase because that is cool
edges.columns = [w.lower() for w in edges.columns]
nodes_df.columns = [w.lower() for w in nodes_df.columns]
```

```python
#I make sure the visualization is readable by making the ids provided in the edges.csv into actual labels instead. We know these from nodes.csv.
id_to_label = nodes_df.set_index('id')['label'].to_dict()

edges['source'] = edges['source'].map(id_to_label)
edges['target'] = edges['target'].map(id_to_label)
```

```python
#Here i define my graph g. I do so by using the pandas framework to make a graph from the pands dataframe that was defined earlier. I do so based on the dataset edges.csv, and i define my source-nodes to be the column "source"
#and my target-nodes to "target" very fittingly so. Furthermore we dont really care about the "type" so we leave that out in the edge attributes, but we do care about weight, so we put that in.
g = nx.from_pandas_edgelist(edges, source='source', target='target', edge_attr=['weight'])
```

```python
#In my aspirations of making a cool looking visualization I decided to calculate some cool metrics that can be used to describe the importance of each node. The first one is degree centrality, which is the degree of a node divided by
#the maximum possible degree in the graph as n-1. As this is undirected, the degree is simply the amount of nodes connected to the node. A high value means that many nodes are connected to that node indicating said node to be influential.
centrality_dgr = nx.degree_centrality(g)
#Eigenvector centrality takes it a step further and includes measurements to think of which nodes a node is connected to, i.e. the scores of the nodes that connects to the node. This means that when determing influence in a network
#we also take into account what the importance of nodes when calculating centrality.
centrality_eigen = nx.eigenvector_centrality_numpy(g, weight='weight')
#Betweenness centrality considers how many nodes would have to go through a node to reach other nodes. It scores a node high in centrality if many nodes are connected via said node. The more paths are made via a node the higher score.
centrality_between = nx.betweenness_centrality(g, weight='weight')
```

```python
#I define these as my attributes to include in my later visualization
nx.set_node_attributes(g, centrality_dgr, 'centrality_dgr')
nx.set_node_attributes(g, centrality_eigen, 'centrality_eigen')
nx.set_node_attributes(g, centrality_between, 'centrality_between')
```

```python
#To make my graph even cooler I would like to add a clustering elements which i do via community detection using the louvain method as this is widdenly used. Simplified this creates a small groups of nodes, starting with each node as 
#its own community. It then merges communities by calculating modularity. If the modularity would increase then the communities are merged, which is then done untill we can no longer merge more communities.
partition = community_louvain.best_partition(g)
#I add the community labels to the nodes
nx.set_node_attributes(g, partition, 'community')
#I define these
nodes = pd.DataFrame.from_dict(dict(g.nodes(data=True)),orient='index')
#I show the top 5s of each community based on eigenvector centrality score
nodes.groupby('community')['centrality_eigen'].nlargest(5)
```

```python
#I decide the spring layout because thats what we did in class
g_layout = nx.layout.spring_layout(g)
#I define plot the graph and add hover tools to show the many attributes i have made and also add coloring so that nodes are colored by their community.
g_plot = hv.Graph.from_networkx(g, g_layout).opts(tools=['hover'], node_color='community')
labels = hv.Labels(g_plot.nodes, ['x', 'y'], 'index')
```

```python
#We visually group the similar nodes to make a readable graph
bundled = bundle_graph(g_plot)
```

```python
#The graph shows a lot of nodes in the central parts of the network, but a lot outside of the central parts only connected to few nodes as well. The community part looks a bit weird, as the clusters of communities dont look that tight
#as there are nodes that have been labeled part of a community but is visually not really grouped with them. 
show(hv.render(bundled * labels.opts(text_font_size='6pt', text_color='white', bgcolor='gray')))
```

### 2. Graph Analysis
Determinate the basic graph indicators:
**Total number of nodes and edges**: to understand the *scale and complexity* of the graph
**Average degree**: giving an average of *connections per node*
**Graph density**:  providing an insight into the *overall connectivity*

```python
# a basic information about the graph
print(f"Graph info:")
print(f"Number of nodes: {g.number_of_nodes()}")
print(f"Number of edges: {g.number_of_edges()}")
print(f"Graph density: {nx.density(g)}") # how close a graph is to being a complete graph
print(f"Graph is directed: {g.is_directed()}")
print(f"Average degree: {sum(dict(g.degree()).values()) / g.number_of_nodes():.2f}")
```

```python
#There is a total number of 70 nodes, meaning that there are 70 characters. These have 299 connections between them as is given by the number of edges 299. This means that there is a lot of connectivity inbetween the nodes, as we have a
#substantial amount of edges compared to nodes. The density of 0.12 is rather low, which means that only 12% of possible connections between nodes exist, but we still have an average degree of 8.5 meaning that on average a node is connected
#to 8.5 other nodes. This leads one towards thinking that some few nodes have a lot of connections between them while most nodes have only a few connections between them. Very fitting when we know that is is a movie we are analysing,
#as one could expect that only a few characters are central to the story and in many scenes, while a lot of other characters only show up once in an interaction with the main characters.
```

### 3. Node analysis
Calculate basic centralities and compare the most central characters.

```python
#I plot the distribution of weights of all edges, to look into whether or not we have more central nodes than others, as it is suspected that we have. 
#To build on that previous idea we can see that majority of edges have a low weight between them, while a few have substantially more. This indicates that some nodes are more important than others.
weights = [z['weight'] for x,y,z in g.edges(data=True)]
sns.displot(weights)
```

```python
#I list the sum of each id (now character name) and their total weight scores. Here we can see that in particular ROSE scores way above the others. We can also see JACK substantially larger while CAL is close to him.
edges_melt = pd.melt(edges, id_vars=['id','weight'], value_vars=['source', 'target'],
        var_name='role', value_name='name')
edges_melt.groupby('name').weight.sum().sort_values(ascending=False)
```

```python
#I got a head of myself a bit earlier so I have already exlpained the meaning of each of these scores, but basically rose is connected to 67% of nodes as is shown by the centrality degree, meaning a big centrality. 
#If we then take into account the importance of her connected nodes her score is also high indicating that the nodes she is connected with are also highly scored. Finally the betweenness score shows that she plays a role also of 
#connecting other nodes.

g.nodes(data=True)['ROSE']
```

### 4. Node attributes
Add Gender attribute (from dataset to the network) and calculate and/or visualize the share of females and males characters in the film. Furthermore, subgraph network to only male characters. Who are the main characters, what is their relationship?



```python
#I add the gender label to my node attributes to visualize in the graph
label_to_gender = nodes_df.set_index('label')['gender'].to_dict()
for node in g.nodes():
    if node in label_to_gender:
        g.nodes[node]['gender'] = label_to_gender[node]
```

```python
#This is the same as earlier but now with gender labels.
g_layout = nx.layout.spring_layout(g)
g_plot = hv.Graph.from_networkx(g, g_layout).opts(tools=['hover'], node_color='community')
labels = hv.Labels(g_plot.nodes, ['x', 'y'], 'index')
bundled = bundle_graph(g_plot)
show(hv.render(bundled * labels.opts(text_font_size='6pt', text_color='white', bgcolor='gray')))
```

```python
#I calculate the counder count and percentage distrubtuion. SO majority of characeters are male.
gender_counts = pd.Series(nx.get_node_attributes(g, 'gender')).value_counts()
gender_percentage = gender_counts / gender_counts.sum() * 100
gender_distribution_df = pd.DataFrame({'Count': gender_counts, 'Percentage': gender_percentage})
print(gender_distribution_df)
```

```python
#I make a subgraph of only the male nodes now.
#What this does is that it seperates the network into three networks instead. Knowing nothing of the movie I cant tell you in too much detail why this is the case, but one could probably guess that leaving out ROSE, the most central character,
#probably means that some nodes are no longer connected even via other nodes.
male_nodes = [node for node, attr in g.nodes(data=True) if attr.get('gender') == 'M']
male_subgraph = g.subgraph(male_nodes)
male_layout = nx.layout.spring_layout(male_subgraph)
male_g_plot = hv.Graph.from_networkx(male_subgraph, male_layout).opts(tools=['hover'], node_color='community')
male_labels = hv.Labels(male_g_plot.nodes, ['x', 'y'], 'index')
male_bundled = bundle_graph(male_g_plot)
show(hv.render(male_bundled * male_labels.opts(text_font_size='6pt', text_color='white', bgcolor='gray')))
```

```python
#Recalculating the centrality measures now only on the male subgraph its still JACK and CALL that are the most important nodes given their respective scores.
degree_centrality_male = nx.degree_centrality(male_subgraph)
eigenvector_centrality_male = nx.eigenvector_centrality(male_subgraph)
betweenness_centrality_male = nx.betweenness_centrality(male_subgraph)

for node in male_subgraph.nodes:
    male_subgraph.nodes[node]['centrality_dgr'] = degree_centrality_male[node]
    male_subgraph.nodes[node]['centrality_eigen'] = eigenvector_centrality_male[node]
    male_subgraph.nodes[node]['centrality_between'] = betweenness_centrality_male[node]

for measure in ['centrality_dgr', 'centrality_eigen', 'centrality_between']:
    top_5 = pd.Series({node: attr[measure] for node, attr in male_subgraph.nodes(data=True) if measure in attr}).nlargest(5)
    print(f"\nTop 5 Male Characters by {measure.replace('_', ' ').title()}:")
    print(top_5)
```

```python
#To gather some information about the two we can look at if they are connected and what the weight of that connection is.
node1 = 'JACK'
node2 = 'CAL'

if male_subgraph.has_edge(node1, node2):
    edge_data = male_subgraph.get_edge_data(node1, node2)
    print(f"Nodes {node1} and {node2} are directly connected.")
    print("Edge Data:", edge_data)
else:
    print(f"Nodes {node1} and {node2} are not directly connected.")


#We can see that the two characters are directly connected, meaning they appear at the same time and the weight tell us it happens 4 times, which is below average actually, indicating that it is not a strong connection.
```
